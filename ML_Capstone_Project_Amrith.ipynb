Bivariate Analysis- we can use heatmap, crosstab, chi-square test

Since all the variables are categorical- we use **Chi square test** and **heatmap**.

The Chi squ test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.

```python
import scipy.stats as stats
```

Two way Tables- use pd.crosstab()- it is used to get the frequency distribution of Corona with respect to other categorical variables.

Bivariate Analysis of **Cough_symptoms** and **Corona**.

Null Hypothesis: There is no relation between categorical variables.

```python
df_observed_cough=pd.crosstab(index=df['Corona'],columns=df["Cough_symptoms"])   #Actual Observations
```

```python
df_observed_cough
```

```python
#Observed Values
Corona_cough_observed_values=df_observed_cough.values
print("Corona_Cough_observed:-\n",Corona_cough_observed_values)
```

```python
val_corona_cough=stats.chi2_contingency(df_observed_cough)
```

```python
val_corona_cough
```

```python
Corona_cough_expected_values=val_corona_cough[3]
```

```python
no_of_rows=len(df_observed_cough.iloc[0:3,0])
no_of_columns=len(df_observed_cough.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degrees of Freedom:-",ddof)
alpha=0.05
```

```python
chi_square=sum([(o-e)**2./e for o,e in zip(Corona_cough_observed_values,Corona_cough_expected_values)])
print(chi_square)
```

```python
chi_square_statistic=chi_square[0]+chi_square[1]
print("Chi-square statistic:",chi_square_statistic)
```

```python
#p-value
from scipy.stats import chi2
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
```

```python
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
sns.heatmap(pd.crosstab(df['Corona'],df['Cough_symptoms']),annot=True)
```

**1. Decision Tree Classifier**

```python
import time

start_time=time.time()
#Seperate the features
X=df[['Cough_symptoms',"Fever","Sore_throat","Shortness_of_breath","Headache","Sex","Age_60_above"]]
y=df["Corona"]

#Convert the categorical variable into Numerical Data using One shot encoding
X=pd.get_dummies(X,columns=['Cough_symptoms',"Fever","Sore_throat","Shortness_of_breath","Headache","Sex","Age_60_above"],drop_first=True)

#Split the data into Training and Testing Sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=True)


from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit(X_train,y_train)


#Make predictions on testing data
y_pred_test=model.predict(X_test)
y_pred_train=model.predict(X_train)

#Evaluation
from sklearn.metrics import accuracy_score
print("Train Accuracy:", accuracy_score(y_train,y_pred_train))
print("Test Accuracy:",accuracy_score(y_test,y_pred_test))
end_time = time.time()

elapsed_time = end_time - start_time

print(f"Elapsed time: {elapsed_time} seconds")
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X_train,y_train,cv=5).mean()  #Cross Validation score

print("Cross Validation Score:",cross_val_score(model,X_train,y_train,cv=5).mean() )
```

```python
import duckdb
import pandas as pd
```

Again the data set was uploaded and cleaned without deleting the "Test_date" using pandas because in the questions "Test_date" columns was required.

```python
df=pd.read_csv("corona_tested_006.csv",index_col=0, na_values='None')
```

```python
df['Cough_symptoms']=df['Cough_symptoms'].fillna(df['Cough_symptoms'].mode()[0])
df['Fever']=df["Fever"].fillna(df["Fever"].mode()[0])
df['Sore_throat']=df["Sore_throat"].fillna(df["Sore_throat"].mode()[0])
df['Shortness_of_breath']=df["Shortness_of_breath"].fillna(df["Shortness_of_breath"].mode()[0])
df["Headache"]=df["Headache"].fillna(df["Headache"].mode()[0])
df["Age_60_above"]=df["Age_60_above"].fillna(df["Age_60_above"].mode()[0])
df["Sex"]=df["Sex"].fillna(df["Sex"].mode()[0])
```

```python
df['Test_date'] = pd.to_datetime(df['Test_date'],format='%d-%m-%Y')
```

```python
conn=duckdb.connect()
conn.register('df',df)
```

```python
df.head(5)
```

**1.Find the number of corona patients who faced shortness of breath.**

```python
conn.execute("select count(*) as No_of_covid_patients_with_breathshortness from df where Corona='positive' and Shortness_of_breath='True'").fetchdf()
```

**2. Find the number of negative corona patients who have fever and sore_throat.**

```python
conn.execute("select count(*) as No_of_covid_patients_fever_sorethroat from df where Corona='negative' and Fever='True' and Sore_throat='True'").fetchdf()
```

**3.Group the data by month and rank the number of positive cases.**


```python
conn.execute("SELECT EXTRACT(MONTH FROM Test_date) AS month, COUNT(*) AS positive_cases_count, RANK() OVER (ORDER BY COUNT(*) DESC) AS case_rank FROM df WHERE Corona = 'positive' GROUP BY EXTRACT(MONTH FROM Test_date) ORDER BY month desc;").fetchdf()
```

**4.Find the female negative corona patients who faced cough and headache.**


```python
conn.execute(" select count(*) as No_of_female_patients_cough_headache from df where Sex='female' and Corona='negative' and Cough_symptoms='True' and Headache='True'").fetchdf()
```

**5.How many elderly corona patients have faced breathing problems?**

```python
conn.execute(" select count(*) as No_of_elderly_patients_breathing_problem from df where Age_60_above='Yes' and Shortness_of_breath='True'").fetchdf()
```

**6.Which three symptoms were more common among COVID positive patients?**

```python
conn.execute("SELECT Three_common_symptom, COUNT(*) AS symptom_count FROM (SELECT CASE WHEN Sore_throat = 'True' THEN 'Sore Throat' WHEN Cough_symptoms = 'True' THEN 'Cough' WHEN Fever = 'True' THEN 'Fever' WHEN Shortness_of_breath = 'True' THEN 'Shortness of Breath' WHEN Headache = 'True' THEN 'Headache' END AS Three_common_symptom FROM df WHERE Corona = 'positive') AS covid_positive_symptoms GROUP BY Three_common_symptom ORDER BY symptom_count DESC LIMIT 3; ").fetchdf()
```

**7.Which symptom was less common among COVID negative people?**

```python
conn.execute("SELECT Less_common_symptom, COUNT(*) AS symptom_count FROM (SELECT CASE WHEN Sore_throat ='True' THEN 'Sore Throat' WHEN Cough_symptoms = 'True' THEN 'Cough' WHEN Fever = 'True' THEN 'Fever' WHEN Shortness_of_breath = 'True' THEN 'Shortness of Breath' WHEN Headache = 'True' THEN 'Headache' END AS Less_common_symptom FROM df WHERE corona ='negative') AS covid_negative_symptoms GROUP BY Less_common_symptom ORDER BY symptom_count ASC LIMIT 1;").fetchdf()
```

**8.What are the most common symptoms among COVID positive males whose known contact was abroad?**

```python
conn.execute("SELECT Common_symptoms_covid_males, COUNT(*) AS symptom_count FROM ( SELECT CASE WHEN Sore_throat = 'True' THEN 'Sore Throat' WHEN Cough_symptoms = 'True' THEN 'Cough' WHEN Fever = 'True' THEN 'Fever' WHEN Shortness_of_breath = 'True' THEN 'Shortness of Breath' WHEN Headache = 'True' THEN 'Headache' END AS Common_symptoms_covid_males FROM df WHERE Corona = 'positive' AND Sex = 'male' AND Known_contact = 'Abroad') AS covid_positive_male_abroad GROUP BY Common_symptoms_covid_males ORDER BY symptom_count DESC; ").fetchdf()
```
